```python
import pandas as pd

details = pd.DataFrame(df[["user_followers","user_friends","user_favourites","user_location"]])
  

details = details.apply(lambda x : True
            if x['user_followers'] <1 and x['user_favourites']<1  else False, axis = 1)
  
# Count number of True in the series
num_rows = len(details[details == True].index)
  
print('Cuentas sin seguidores ni usuarios seguidos : ',
      num_rows )
```
